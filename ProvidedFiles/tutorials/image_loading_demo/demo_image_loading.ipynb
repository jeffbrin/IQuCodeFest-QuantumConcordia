# Démonstration - Chargement d'Images

Démonstration rapide du chargement et de l'affichage d'images.

````python
# Load and display a binary image (black/white)
#binary_image = load_binary_image(IMAGE_BINARY, size=DEFAULT_SIZE, threshold=128)
binary_image = load_binary_image(IMAGE_PATH, size=DEFAULT_SIZE, threshold=128)

print("Binary Image Info:")
print(f"  Shape: {binary_image.shape}")
print(f"  Values: 0 (white) and 1 (black)")
print(f"  Black pixels: {binary_image.sum()}/{binary_image.size}")

display_image(binary_image, title=f"Binary Image ({DEFAULT_SIZE}x{DEFAULT_SIZE})", cmap='binary')
````

## Référence Rapide

Utilisez ces fonctions dans vos notebooks :

```python

# Chargement d'images
binary_img = load_binary_image("images/votre_image.png", size=32, threshold=128)
gray_img = load_grayscale_image("images/votre_image.png", size=64, normalize=True)
color_img = load_color_image("images/votre_image.png", size=64, normalize=True)

# Affichage

```

### Paramètres :
- **size** : Dimensions de l'image (ex: 32, 64, 128) - les puissances de 2 fonctionnent mieux pour les circuits quantiques
- **threshold** : Pour les images binaires (0-255), les pixels en dessous du seuil deviennent noirs (1)
- **normalize** : Si True, normalise les valeurs dans l'intervalle [0, 1] (utile pour l'encodage quantique)
- **background_color** : Pour les images transparentes, utilisez 'white', 'black', ou un tuple RGB
- **cmap** : Carte de couleurs pour l'affichage ('gray', 'binary', 'gray_r' pour inversé, etc.)
